### 1. Introduction:

One of the primary determinants of an individual’s quality of health is related to the potential spatial access to healthcare (Pecurra, Piacenza and Turati, 2018). In relation urban planning within the United Kingdom a central consideration is to ensure universal access to healthcare. The aim of this project is to assess the spatial inequality of healthcare within Northern Ireland and examine potential new sites for general hospitals. 

This project maintains substantial importance as Northern Ireland hospital waiting lists are the worst regionally across the UK with 428,858 patients on a waiting list for their initial consultant-led outpatient appointment as of September 2023 (NISRA, 2024), reflecting the struggling healthcare system. Further, Northern Ireland has a significant rural population with 42.5% of those serviced by the Northern Health and Social Care Trust, living in rural areas. Ensuring those living within rural areas have adequate access to healthcare services is a key challenge and consideration (NHSCT, 2024).

Adopting a Northern Ireland context, one can acknowledge the potential knowledge gap academically in relation to analysis on healthcare inequality spatially which this study aims to address. One such study which is an exception is Jo, Kruger, and Tennant’s (2020) geospatial analysis on urban and rural access to dental services in Northern Ireland. Furthermore, in relation to utilizing Multi-Criteria Evaluations (MCE) to explore possible suitable locations for healthcare facilities, similar studies have been conducted by Zandi et al., (2024) which assessed hospital site suitability in Tehran, and by Dutta et al., (2021) which employed spatial analysis to model optimal locations for healthcare locations in West Bengal.

This report will initially employ QGIS to consider hospital proximity, population density and land suitability, through which an MCE will be employed to aid in recognising the most optimal locations for new hospital development. Utilizing Python coding, the results are analysed and visualised, identifying the ten most optimal locations for new hospital developments.

*2.4. Method 2: Population Distribution* 


To understand the population distribution throughout Northern Ireland a choropleth map was created to illustrate population density, a crucial consideration as reflected through being a criterion within a study on optimal hospital location in Shiraz conducted by Rahimi, Goli, and Rezaee (2016) given a weight of 0.1986. This was achieved through joining two csv files, one of which indicated geometry and the second one providing population density. The joined layer was then rasterized, normalized and through employing a singleband pseudocolour graduated scale the map was created. Illustrating population density will aid in understanding the largest population centres and through overlaying hospitals locations it is possible to identify unserviced areas.

*2.5. Method 3: Land Suitability Multi-Criteria Evaluation*


As a core element within Method 4 and the overall aim of the study, understanding areas which are suitable for hospital construction is vital. Through incorporating five primary factors and constraints, a suitability index was created. Each factor and constraint were rasterized and then normalised to enable the MCE to be conducted utilizing the Raster Calculator. Through analysis of a similar study conducted by Almansi et al. (2022) which explored hospital site suitability in Malacca, Malaysia utilizing Analytical Hierarchy Process (APH) Models, a series of weights were decided upon for this study, adjusted to represent the geographical difference in location and specific needs within Northern Ireland. APH Models are the most widely employed strategy within Multiple-Criteria Decision Analysis (MCDA) academic studies (Almansi et al., 2022). This involves understanding the research project, identifying criteria which reflect the needs of the project, pair-wise comparison through ranking based on importance based on the specific parameters and considerations, and then applying weights based on the model (Vaidya and Kumar, 2006). This project identified Evelvation, Slope, Proximity to Roads, Protected Areas, and Land Cover as primary criteria. Protected Areas and Land Covers were constraints within this context, Proximity to Roads was identified as being an important consideration and was given a weight of 0.55 and both Slope and Elevation were given a weight of 0.225 based on importance to the suitability within the project and based upon other scholarly approaches. Considering land suitability for potential hospital sites provides greater depth to the study, reinforcing any suggestions made. 

*2.6. Method 4: Multi-Criteria Evaluation for Optimal Hospital Location*

Method 4 involves bringing together the three previous layers of analysis to understand optimal locations for a new hospital site in Northern Ireland. To reflect the inclusion of proximity to the nearest hospital, population density and land suitability, the weights within the MCE were adjusted to reflect the needs of the area of study, taking inspiration from similar studies and employing an APH model. The Land Suitability Raster created within Method 3 was given a weight of 0.4 , and both proximity to the nearest hospital and population density receiving a weight of 0.3.

Considering both proximity to elevation and population ensure that any new hospital sites are located in areas where there remains both a need spatially and a demand whilst including land suitability highlights suitable locations for development within this area. The importance of each factor is reflected within the weights attributed to each factor.

### 4. Interpretation of Results and Discussion

Looking initially at Figure 2 one can note the Eastern portion of Northern Ireland is relatively well serviced, with a potential exception in the South-East, whilst there is a perceived lack of hospitals located within the Western Health and Social Care Trust. The potential spatial inequality of healthcare in proximity to the Western and South-South-Western borders can be acknowledged. However, through examining Figure 3, these observations are contextualised through acknowledging the population density often reflects location of existing hospitals. This links to a likely consideration within urban planning of aiming to place hospitals in central areas of population hubs to maximise its sphere of influence and it is unlikely to place hospitals near border areas absent of significant populations. This can be seen through MCE analysis conducted by Zandi et al., (2024) within Tehran which placed the need to be located near considerable populations as the fourth highest weight of eleven criterion, reflecting the importance of this as a criterion within urban planning.

Briefly touching upon Figure 4 which explore land suitability, one can note the abundance of potential suitable sites which reflects regional areas which are more rural, and agriculture as a likely major economic output, potentially having implications on lifestyle and interactions with health systems.

Finally, in relation to Figure 5, Figure 6 and Table 2 one can reiterate the proximity to border areas within Western and Southern regions. Through identifying the top ten most optimal locations for new hospital locations one can note key areas which are underserviced across a range of data zones. There is a notable cluster of four locations which suggests high suitability and need for improved access to hospital facilities in proximity to that data zone and the town of Castlederg. An outlier is observed in the South-East; however, this location would likely not be feasible due to the imposed Mourne Mountains in the region which will lower its sphere of influence. This need to contextualise locations is required across all ten identified areas. It is possible that the proximity weighting attributed has overly influenced results and in relation to the cluster near Castlederg it is likely there would not be a sufficient population, seen through Figure 3, to warrant hospital construction. On reflection a population threshold in proximity of possible locations could have been applied.

On balance, it is likely not feasible to construct new hospital sites in these border locations identified and in fact a conclusion can be reached that Northern Ireland is relatively well serviced in relation to hospitals. However, there remains value in these findings as identifying areas which are experiencing spatial inequality in relation to healthcare such as Castlederg provides a foundation for further study and governmental review examining whether these areas maintain acceptable levels of primary care and frameworks to ensure that the spatial inequality experienced in rural areas is minimised; a standpoint influenced by a review of primary care in Northern Ireland by Connolly et al., (2022).

```python
optimal_arr = optimal.read(1) #loading raster data
optimal_arr
```

```python
image_read_masked = np.ma.masked_array(optimal_arr, mask=(optimal_arr <= 0)) #masking values
```

```python
# Find the top coords based on values
# argsort sorts array, gets top 10000, then returns index of each of the top values
top_10_indices = np.argsort(optimal_arr.flatten())[-10000:][::-1]
```

```python
# go from flattened index back to 2d rows and cols representation
rows, cols = np.unravel_index(top_10_indices, image_read_masked.shape)
```

```python
# distance between 2 x-y points
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
# saving row and col points far enough away from other points
selected_rows = [] 
selected_cols = [] 

# loop over row values
for i in range(len(rows)):
    is_far = True
    
    # loop over already selected points
    for j in range(len(selected_rows)):
        # if point is too close to another point, ignore it
        if euclidean_distance(rows[i], cols[i], selected_rows[j], selected_cols[j]) < 100:
            is_far = False

    # if far enough, select it as a point
    if is_far:
        selected_rows.append(rows[i])
        selected_cols.append(cols[i])
    
    if len(selected_rows) == 10:
        print(i)
        break

selected_rows
```
